```python
'''
Q1. Discuss string slicing and provide examples.

A1: String slicing is a technique in Pythonvthat allows you to extract a portion of a string from a given original string.

Basic syntax:
string_variable[start_index:end_index:step]

'''

#Example:-
string = "Hello, World!"
substring = string[7:12]
print(substring)  # Output: World

#Example:-
string = "Python is a programming language"
substring = string[0:6]
print(substring)  # Output: Python

#Example:-
string = "Python is a programming language"
substring = string[0:6:2]
print(substring)  # Output: Pto

#Example:-
string = "Python is a programming language"
substring = string[0:6:-1]
print(substring)  # Output: None

#Example:-
string = "Python is a programming language"
substring = string[::-1]
print(substring)  # Output: egaugnal a si nohtyP

#Example:-
string = "Python is a programming language"
substring = string[::2]
print(substring)  # Output: Pto tigl
```

```python
'''
Q2. Explain the key features of lists in Python.

A2. Lists are fundamental data structure that allowws you to store and manipulate a collection of items.

Some key features of lists in Python are:-
'''

#1. Ordered:-
my_list = [1, 2, 3, 4, 5]
print(my_list)

#2. Mutable:-
my_list = [1, 2, 3, 4, 5]
my_list[0] = 10
print(my_list)

#3. Dynamic:-
my_list = [1, 2, 3, 4, 5]
my_list.append(6)
print(my_list)

#4. Heterogeneous:-
my_list = [1, "hello", 3.14, True]
print(my_list)

#5. Nested:-
my_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(my_list)

#6. Slicing:-
my_list = [1, 2, 3, 4, 5]
subset = my_list[1:4]
print(subset)

#7. Iteration:-
my_list = [1, 2, 3, 4, 5]
for item in my_list:
    print(item)

#8. Concatenation:-
list1 = [1, 2, 3]
list2 = [4, 5, 6]
concatenated_list = list1 + list2
print(concatenated_list)

#9. Length:-
my_list = [1, 2, 3, 4, 5]
length = len(my_list)
print(length)

#10. Sorting:-
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
sorted_list = sorted(my_list)
print(sorted_list)

#11. Reversing:-
my_list = [1, 2, 3, 4, 5]
reversed_list = my_list[::-1]
print(reversed_list)

#12. Copying:-
original_list = [1, 2, 3, 4, 5]
copied_list = original_list.copy()
print(copied_list)

#13. Membership:-
my_list = [1, 2, 3, 4, 5]
print(3 in my_list)  # Output: True
print(6 in my_list)  # Output: False

#14. Index:-
my_list = [1, 2, 3, 4, 5]
index = my_list.index(3)
print(index)  # Output: 2

#15. Count:-
my_list = [1, 2, 3, 4, 5, 3, 3, 3]
count = my_list.count(3)
print(count)  # Output: 3

#16. Clear:-
my_list = [1, 2, 3, 4, 5]
my_list.clear()
print(my_list)  # Output: []
```

```python
'''
Q3. Describe how to access, modify, and delete elements in a list with examples.

A3. In Python, you can access, modify, and delete elements in a list using various methods and operations.
'''

#1. Accessing elements:- You can access the elements of list using indices. Indices start at 0 for the first element and go up to len(list) - 1 for the last element. Negative indices can be used to access elements from the end of the list, with -1 being the last element.

#Example:-

lis = [1, 2, 3, 4, 5]
print(lis[0])  # Output: 1
print(lis[-1])  # Output: 5
print(lis[0:3])  # Output: [1, 2, 3]

#2. Modifying elements:- You can modify elements in list by assigning new values to a specific index.

#Example:-

lis = [1, 2, 3, 4, 5]
lis[0] = 10
print(lis)  # Output: [10, 2, 3, 4, 5]

#3. Deleting elements:- You can delete elements from list using the del keyword, the pop() method, the remove() method and clear method.

#Example:-

lis = [1, 2, 3, 4, 5]
del lis[0]
print(lis)  # Output: [2, 3, 4, 5]

lis = [1, 2, 3, 4, 5]
popped_element = lis.pop(2)
print(lis)  # Output: [1, 2, 4, 5]
print(popped_element)  # Output: 3

lis = [1, 2, 3, 4, 5]
lis.remove(3)
print(lis)  # Output: [1, 2, 4, 5]

lis = [1, 2, 3, 4, 5]
lis.clear()
print(lis)  # Output: []
```

```python
'''
Q4.  Compare and contrast tuples and lists with examples.

A4. Tuples and lists are both data structures in Python that are used to store collections of items. However, there are some key differences between them that make them different.

Similarities:-

1. Ordered:- Both list and tuple maintain the order of the elements.
2. Heterogeneous Elements:- Both list and tuple can contain elements of different data types.
3. Indexing & Slicing:- Both list and tuple can be indexed and sliced.
4. Iteration:- Both list and tuple can be iterated
'''
#Differences:-

#Mutability:- Lists are mutable, meaning their elements can be modified, added, or removed.
             #Tuples are immutable, meaning their elements cannot be modified once they are assigned.

#Example:-
lis = [1, 2, 3, 4, 5]
lis[0] = 10
print(lis)  # Output: [10, 2, 3, 4, 5]

tup = (1, 2, 3, 4, 5)
tup[0] = 10  # This will raise a TypeError

#Syntax:- Lists are defined using square brackets [].
         #Tuples are defined using parentheses ().

#Example:-
lis = [1, 2, 3, 4, 5]
tup = (1, 2, 3, 4, 5)

#Performance:- Tuples are generally faster than lists.

#Methods & Operations:- Lists have more built-in methods and operations than tuples.
                       #Tuples have fewer built-in methods and operations

#Example:-
lis = [1, 2, 3, 4, 5]
lis.append(6)
print(lis)  # Output: [1, 2, 3, 4, 5, 6]

tup = (1, 2, 3, 4, 5)
tup.append(6)  # This will raise a AttributeError

#Example:-
lis = [1, 2, 3, 4, 5]
lis.sort()
print(lis)  # Output: [1, 2, 3, 4, 5]

tup = (3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5)
sorted_tup = sorted(tup)
print(sorted_tup)

#Example:-
lis = [1, 2, 3, 4, 5]
lis.extend([6, 7, 8])
print(lis)

tup = (1, 2, 3, 4, 5)
tup.extend([6, 7, 8])  # This will raise a AttributeError

#Example:-
lis = [1, 2, 3, 4, 5]
del lis[0]
print(lis)  # Output: [2, 3, 4, 5]

tup = (1, 2, 3, 4, 5)
del tup[0]  # This will raise a TypeError
```

```python
'''
Q5. Describe the key features of sets and provide examples of their use.

A5. Sets are a built-in data structure in Python that represents an unordered collection of unique elements.

Some key features of sets are:-
'''

#1. Unordered:-
my_set = {3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5}
print(my_set)

#2. Mutable:-
my_set = {3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5}
my_set.add(7)
print(my_set)

#3. Heterogeneous:-
my_set = {1, "hello", 3.14, True}
print(my_set)

#4. No Duplicates:-
my_set = {3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5}
print(my_set)

#5. Iteration:-
my_set = {3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5}
for item in my_set:
    print(item)


#6. Membership:-
my_set = {3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5}
print(3 in my_set)
print(6 in my_set)

#7. Copying:-
original_set = {1, 2, 3, 4, 5}
copied_set = original_set.copy()
print(copied_set)

#8. Length:-
my_set = {3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5}
length = len(my_set)
print(length)

#9. Set Operations:-

#Union:- Combine elements from both sets (All unique elements).
#Example:-
s1 = {1, 2, 3}
s2 = {3, 4, 5}
union_set = s1.union(s2)
print(union_set)

#Intersection:- Get elements that are common in both sets.
#Example:-
s1 = {1, 2, 3}
s2 = {3, 4, 5}
intersection_set = s1.intersection(s2)
print(intersection_set)

#Set Difference:- Get elements present in the first set and not in second.
#Example:-
s1 = {1, 2, 3}
s2 = {3, 4, 5}
difference_set = s1.difference(s2)
print(difference_set)
```

```python
'''
Q7. Describe how to add, modify, and delete items in a dictionary with examples.

A7. In Python, you can add, modify, and delete items in a dictionary which are mutable using various methods and operations.
'''

#Adding Items to dictionary:- You can add items to a dictionary by specifying a new key and assigning a value to it.

#Example:-
my_dict = {'a': 1, 'b': 2, 'c': 3}
my_dict['d'] = 4
print(my_dict)

#Modifying Items in a dictionary:- You can modify the value of an existing item in a dictionary by assigning a new value to it.

#Example:-
my_dict = {'a': 1, 'b': 2, 'c': 3}
my_dict['a'] = 10
print(my_dict)

#Deleting Items from a dictionary:- You can delete items from a dictionary using the del keyword, the pop() method, or the popitem() method.

#Example:-
my_dict = {'a': 1, 'b': 2, 'c': 3}
del my_dict['a']
print(my_dict)
```
